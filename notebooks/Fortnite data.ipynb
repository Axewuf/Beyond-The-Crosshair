```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
import json
import ast

alt.data_transformers.enable("default", max_rows=None)
alt.renderers.enable("mimetype")
alt.theme.enable("fivethirtyeight")
```

## Datasets & Cleaning

### Fortnite

```python
fn_data = pd.read_csv("../assets/raw/GG Fortnite.csv")
```

```python
fields_to_keep = ['month', 'peak', 'gain', '% gain']

fn_data.columns = fn_data.columns.str.lower()
fn_data = fn_data.replace("-", "0").replace("—", "0") # Replacing both the short - and long —
fn_data[['peak', 'gain', '% gain']] = fn_data[['peak', 'gain', '% gain']].replace({",": "", "%": ""}, regex=True).astype(float)
fn_data = fn_data[fields_to_keep][1:] # Getting rid of first row of Last 30 days
fn_data["month"] = pd.to_datetime(fn_data["month"], format="%B %Y")
fn_data = fn_data.sort_values("month")
fn_data = fn_data[:-1]
fn_data
```

```python
fn_data.to_csv("../assets/clean/GG_Fortnite_Clean.csv", encoding="utf-8", index=False, header=True)
```
